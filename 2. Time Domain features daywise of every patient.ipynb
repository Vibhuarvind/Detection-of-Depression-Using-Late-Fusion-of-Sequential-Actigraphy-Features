```python
import pandas as pd
import numpy as np

data_path = 'data - depression/'

# Process individual condition files
for i in range(1, 24):
    my_file = data_path + 'condition/condition_' + str(i) + '.csv'
    df_cond = pd.read_csv(my_file)

    # group by date of condition.csv
    df_cond_by_date = df_cond.groupby(['date'], as_index=False).agg(
        n=pd.NamedAgg(column='activity', aggfunc='count'),
        mean_act=pd.NamedAgg(column='activity', aggfunc='mean'),
        rms_act=pd.NamedAgg(column='activity', aggfunc=lambda x: np.sqrt(np.mean(x**2))),
        std_act=pd.NamedAgg(column='activity', aggfunc='std'),
        skewness_act=pd.NamedAgg(column='activity', aggfunc='skew'),
        kurtosis_act=pd.NamedAgg(column='activity', aggfunc=lambda x: x.kurtosis()),
        shape_factor_act=pd.NamedAgg(column='activity', aggfunc=lambda x: np.mean(x**2) / (np.mean(x)**2)),
        crest_factor_act=pd.NamedAgg(column='activity', aggfunc=lambda x: np.max(x) / np.sqrt(np.mean(x**2))),
        impulse_factor_act=pd.NamedAgg(column='activity', aggfunc=lambda x: np.max(np.abs(x))),
        peak_activity=pd.NamedAgg(column='activity', aggfunc='max'),
        variance_act=pd.NamedAgg(column='activity', aggfunc='var'),
        autocorrelation_act=pd.NamedAgg(column='activity', aggfunc=lambda x: np.corrcoef(x[:-1], x[1:])[0, 1]),
    )

    # remove incomplete days from stats
    #24*60 = 1440
    df_cond_by_date = df_cond_by_date[df_cond_by_date.n == 1440]

    # Add a new column "Label" with the value "condition"
    df_cond_by_date['Label'] = 'condition'

    # save extracted features to csv file
    df_cond_by_date.to_csv('C:/Users/hp/TDF/condition/condition_' + str(i) + '_features.csv', index=False)

# Concatenate individual condition feature files
all_features = pd.DataFrame()

for i in range(1, 24):
    my_file = 'C:/Users/hp/TDF/condition/condition_' + str(i) + '_features.csv'
    df_cond_features = pd.read_csv(my_file)

    all_features = all_features.append(df_cond_features)

# Add a new column "Label" with the value "condition" to the concatenated DataFrame
all_features['Label'] = 'condition'

# Save the concatenated DataFrame to a new CSV file
all_features.to_csv('TDF_CONDITION.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import glob

data_path = 'data - depression/'

for i in range(1, 33):
    my_file = data_path + 'control/control_' + str(i) + '.csv'
    df_control = pd.read_csv(my_file)

    # group by date of condition.csv
    df_control_by_date = df_control.groupby(['date'], as_index=False).agg(
        n=pd.NamedAgg(column='activity', aggfunc='count'),
        mean_act=pd.NamedAgg(column='activity', aggfunc='mean'),
        rms_act=pd.NamedAgg(column='activity', aggfunc=lambda x: np.sqrt(np.mean(x**2))),
        std_act=pd.NamedAgg(column='activity', aggfunc='std'),
        skewness_act=pd.NamedAgg(column='activity', aggfunc='skew'),
        kurtosis_act=pd.NamedAgg(column='activity', aggfunc=lambda x: x.kurtosis()),
        shape_factor_act=pd.NamedAgg(column='activity', aggfunc=lambda x: np.mean(x**2) / (np.mean(x)**2)),
        crest_factor_act=pd.NamedAgg(column='activity', aggfunc=lambda x: np.max(x) / np.sqrt(np.mean(x**2))),
        impulse_factor_act=pd.NamedAgg(column='activity', aggfunc=lambda x: np.max(np.abs(x))),
        peak_activity=pd.NamedAgg(column='activity', aggfunc='max'),
        variance_act=pd.NamedAgg(column='activity', aggfunc='var'),
        autocorrelation_act=pd.NamedAgg(column='activity', aggfunc=lambda x: np.corrcoef(x[:-1], x[1:])[0, 1]),
    )

    # remove incomplete days from stats
    df_control_by_date = df_control_by_date[df_control_by_date.n == 1440]

    # Add a new column "Label" with the value "control"
    df_control_by_date['Label'] = 'control'

    # save extracted features to csv file
    df_control_by_date.to_csv('C:/Users/hp/TDF/control/control_' + str(i) + '_features.csv', index=False)

# Path to the directory containing the CSV files
folder_path = 'C:/Users/hp/TDF/control'

# Use glob to get a list of all CSV files in the directory
csv_files = glob.glob(f'{folder_path}/*.csv')

# Initialize an empty DataFrame to store the concatenated data
all_data = pd.DataFrame()

# Loop through the list of CSV files and concatenate them
for file in csv_files:
    df = pd.read_csv(file)
    all_data = pd.concat([all_data, df], ignore_index=True)

# Add a new column "Label" with the value "control" to the concatenated DataFrame
all_data['Label'] = 'control'

# Save the concatenated DataFrame to a new CSV file
all_data.to_csv('TDF_CONTROL.csv', index=False)
```

```python
import pandas as pd

def concatenate_csv(file1, file2, output_file):
    # Read the CSV files into pandas DataFrames
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)

    # Concatenate the DataFrames vertically (axis=0 means along the rows)
    concatenated_df = pd.concat([df1, df2], axis=0, ignore_index=True)

    # Write the concatenated DataFrame to a new CSV file
    concatenated_df.to_csv(output_file, index=False)

# Example usage
file1_path = 'TDF_CONDITION.csv'
file2_path = 'TDF_CONTROL.csv'
output_file_path = 'TDF_control_condition.csv'

concatenate_csv(file1_path, file2_path, output_file_path)
```
